```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Pathway of Whale_return, algo_returns, s&p trading history

whale_p = Path ("C:/Users/DELL/Desktop/panda_hw_r/whale_return.csv")
algo_p = Path ("C:/Users/DELL/Desktop/panda_hw_r/algo_returns.csv")
sptsx_p = Path ("C:/Users/DELL/Desktop/panda_hw_r/sp_tsx_history.csv")
otex_p = Path ("C:/Users/DELL/Desktop/panda_hw_r/otex_historical.csv")
shop_p = Path ("C:/Users/DELL/Desktop/panda_hw_r/shop_historical.csv")
```

```python
# read cvsfile of Whale_return, algo_returns, s&p trading history
whale_return_df = pd.read_csv(whale_p, infer_datetime_format=True, parse_dates=True, index_col="Date")
algo_df = pd.read_csv(algo_p, infer_datetime_format=True, parse_dates=True, index_col="Date")
sptsx_df = pd.read_csv(sptsx_p, infer_datetime_format=True, parse_dates=True, index_col="Date")
otex_df = pd.read_csv(otex_p, infer_datetime_format=True, parse_dates=True, index_col="Date")
shop_df = pd.read_csv(shop_p, infer_datetime_format=True, parse_dates=True, index_col="Date")
```

```python
whale_return_df.sort_index(ascending = True, inplace = True)
whale_return_df.head()
```

```python
# Count nulls of Whale_returns
whale_return_df.isnull().sum()
whale_return_df.isnull().mean()*100
```

```python
# Drop nulls of Whale_returns
whale_return_df.dropna(inplace=True)
```

```python
# checking if nulls are removed from Whale_returns
whale_return_df.isnull().mean()*100
```

```python
whale_return_df.dtypes
```

```python
# Reading algorithmic returns
algo_df.sort_index(ascending = True, inplace = True)
algo_df.head()
```

```python
# Count nulls of algo returns
algo_df.isnull().mean()*100
```

```python
# Drop nulls of algo returns
algo_df.dropna(inplace=True)
```

```python
# checking if nulls are removed from Whale_returns
algo_df.isnull().mean()*100
```

```python
# Reading S&P TSX 60 Closing Prices
sptsx_df.sort_index(ascending = True, inplace = True)
sptsx_df.head()
```

```python
sptsx_df.dtypes
```

```python
# Fix Data Types of S&P TSX 60 Closing Prices
sptsx_df["Close"]= sptsx_df["Close"].str.replace('$',"")
sptsx_df["Close"]= sptsx_df["Close"].str.replace(',',"")
sptsx_df["Close"]
```

```python
sptsx_df["Close"].fillna(0)
```

```python
# converting to numeric for sptsx
sptsx_df["Close"]= sptsx_df["Close"].astype(float)
```

```python
sptsx_df.dtypes
```

```python
# Convert the S&P TSX 60 closing prices to daily returns.
sptsx_daily_returns = sptsx_df.pct_change()
sptsx_daily_returns.dropna(inplace = True)
sptsx_daily_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sptsx_daily_returns.rename(columns = {"Close":"sptsx"}, inplace = True)
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
combined_df= pd.concat([whale_return_df,algo_df,sptsx_daily_returns], axis="columns", join="inner")
combined_df
```

```python
# Plot daily returns of all portfolios
combined_df.plot(figsize = (20,10))
```

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + combined_df).cumprod()
cumulative_returns.head() 
# Plot cumulative returns
cumulative_returns.plot(figsize=(10,5)) 
```

```python
message = "Yes, Algo 1's portfolio is leading followed by Berkshire hathaway Inc. Also, Algo 2 and Soros fund management is outperforming sptsx."
print (message)
```

```python
# Box plot to visually show risk
combined_df.plot(kind = "box", figsize = (20,10))
```

```python
# Calculate the daily standard deviations of all portfolios
daily_std_df = pd.DataFrame(combined_df.std()).rename(columns = {0:"Standard_Deviation"})
daily_std_df.head(7)
```

```python
# Determine which portfolios are riskier than the S&P TSX 60
higher_std = daily_std_df[daily_std_df["Standard_Deviation"] > daily_std_df.loc["sptsx", "Standard_Deviation"]]
higher_std
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = daily_std_df* np.sqrt(252)
annualized_std.head(7) 
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window and plot. 
rolling_std = combined_df.rolling(window=21).std().plot(figsize=(20,10)) 
```

```python
# Calculate the correlation
correlation = pd.DataFrame(combined_df.corr())
correlation
```

```python
# Display de correlation matrix
import seaborn as sns
sns.heatmap(correlation, vmin=-1, vmax=1,)
```

```python
# Calculate covariance of a single portfolio
covariance = combined_df['SOROS FUND MANAGEMENT LLC'].cov(combined_df['sptsx'])
covariance
```

```python
# Calculate variance of S&P TSX
variance = combined_df['sptsx'].var()
variance
```

```python
# Calculate beta of all daily returns of SOROS FUND MANAGEMENT LLC
soros_beta = covariance / variance
soros_beta
```

```python
# calculate and plot the 60-day rolling beta for Soros LLC and the S&P TSX 60
rolling_covariance = combined_df['SOROS FUND MANAGEMENT LLC'].rolling(window=60).cov(combined_df['sptsx'])
rolling_covariance = rolling_covariance.dropna()
rolling_variance = combined_df['sptsx'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
```

```python
# Plot beta trend
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of Soros fund management LLC')
```

```python
# Use `ewm` to calculate the rolling window
combined_df.ewm( halflife=21).std().plot(figsize=(20,15))
```

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios") 
```

```python
#Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.
print ("Looking at the bar graph above, Algorithmic strategy outperforms both the market sptsx and whales portfolio")
```
